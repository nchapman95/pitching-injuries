## Feature Engineering

```python
def number_seasons_prior(Player, Season):
    return Season - pitches_injury[pitches_injury['Name'] == Player]['Season'].min()
```

```python
pitches_injury['num_seasons_a_priori'] = [number_seasons_prior(x,y) 
                                 for x,y 
                                 in pitches_injury[['Name', 'Season']].itertuples(index=False)]
```

```python
pitches_injury['seasons_until_injury'] = [x + 1 if y == 1 else 0 for x,y 
                                          in pitches_injury[['num_seasons_a_priori', 'injured_tag']]
                                          .itertuples(index=False)]
```

```python
for pitch_type in pitch_types:
    pitches_injury[pitch_type[:-1] + "_count"] = pitches_injury[pitch_type] * pitches_injury['Pitches']
```

```python
def pitch_diff(Player, Season):
    temp_player = pitches_injury[(pitches_injury['Name'] == Player)
                                &(pitches_injury['Season'] <= Season)
                                & (pitches_injury['Season'] > Season - 3)]
    
    curr_player = pitches_injury[(pitches_injury['Name'] == Player)
                                &(pitches_injury['Season'] == Season)]
    
    temp_df = pd.DataFrame(temp_player[delta_types].mean()).T
    temp_df['Player'] = Player
    temp_df['Season'] = Season
    return temp_df
```

```python
pitches_injury['pitches_per_game'] = pitches_injury['Pitches'] / pitches_injury['G_x']
pitches_injury["IP_per_G"] = pitches_injury['IP_x'] / pitches_injury['G_x']
```

```python
delta_types = []
```

```python
for x in pitch_types:
    delta_types.append(x)
```

```python
for x in pitch_types:
    delta_types.append(x[:-1] + "_count")
    delta_types.append(x[:-1] + "v")
```

```python
delta_types
```

```python
delta_types.append("Pitches")
delta_types.append("G_x")
delta_types.append("GS_x")
delta_types.append("pitches_per_game")
delta_types.append("IP_per_G")
```

```python
pitch_diffs = [pitch_diff(x,y) for x,y in pitches_injury[['Name', 'Season']].itertuples(index=False)]
```

```python
pitch_diffs = pd.concat(pitch_diffs)
```

```python
pitch_diffs.columns = [x + "_3_seasons_prev" if (x != 'Player') and (x != 'Season') else x
                       for x in pitch_diffs.columns]
```

```python
pitches_injury = pd.merge(pitches_injury, pitch_diffs, left_on =['Name','Season'], right_on=['Player','Season'] )
```

```python
for each in delta_types:
    pitches_injury['{}_delta'.format(each)] = pitches_injury[each] - pitches_injury[each + "_3_seasons_prev"]
    
```
